```python
# 33. Crea una funci√≥n lambda que sume elementos correspondientes de dos listas dadas.
from itertools import zip_longest

sumar_listas = lambda lista1, lista2, fillvalue=0: [a + b for a, b in zip_longest(lista1, lista2, fillvalue=fillvalue)]

# Ejemplo
l1 = [1, 2, 3, 7]
l2 = [4, 5]
print(sumar_listas(l1, l2))            # [5, 7, 3, 7]  (3+0=3, 7+0=7)
print(sumar_listas(l1, l2, fillvalue=10))  # si prefieres rellenar con 10
```
