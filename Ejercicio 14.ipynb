```python
# 14. Crea una función que retorne las palabras de una lista de palabras que comience con una letra en especifico. Usa la
# función filter()
def palabras_que_comienzan_con(lista_palabras, letra):
    letra = letra.lower()  # Normalizamos
    return list(filter(lambda palabra: palabra.lower().startswith(letra), lista_palabras))


# Ejemplo de uso:
palabras = ["Casa", "coche", "Perro", "cielo", "Mesa", "Camino"]
print(palabras_que_comienzan_con(palabras, "c"))
```
