```python
#3. Escribe una funci칩n que tome una lista de palabras y una palabra objetivo como par치metros. La funci칩n debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.
def filtrar_palabras(lista_palabras, palabra_objetivo):
    palabra_objetivo = palabra_objetivo.strip().lower()
    resultado = []

    for palabra in lista_palabras:
        palabra_limpia = palabra.strip().lower()
        print(f"Comparando '{palabra_limpia}' con '{palabra_objetivo}'")  # <- Depuraci칩n
        if palabra_objetivo in palabra_limpia:
            resultado.append(palabra)
    
    return resultado


palabras = ["Gato", "perro", "gatito", "tigre", "agotar", "gatota", "gatear"]
objetivo = "gato"
resultado = filtrar_palabras(palabras, objetivo)
print("Resultado final:", resultado)
```
