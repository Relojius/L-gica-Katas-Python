```python
# 41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el
# monto final de una compra en una tienda en línea, después de aplicar un descuento.
# 1. Solicitar el precio original
precio_original = input("Ingresa el precio original del artículo: ")

try:
    precio_original = float(precio_original)
    if precio_original < 0:
        print("El precio no puede ser negativo.")
    else:
        # 2. Preguntar si tiene cupón de descuento
        tiene_cupon = input("¿Tienes un cupón de descuento? (sí/no): ").strip().lower()

        if tiene_cupon == "sí" or tiene_cupon == "si":
            # 3. Solicitar valor del cupón
            cupon = input("Ingresa el valor del cupón de descuento: ")
            try:
                cupon = float(cupon)
                if cupon > 0:
                    precio_final = precio_original - cupon
                    if precio_final < 0:
                        precio_final = 0  # Evitar precio negativo
                    print(f"El precio final después del descuento es: {precio_final:.2f} €")
                else:
                    print("El valor del cupón debe ser mayor a 0. No se aplica descuento.")
                    print(f"El precio final es: {precio_original:.2f} €")
            except ValueError:
                print("Cupón inválido. No se aplica descuento.")
                print(f"El precio final es: {precio_original:.2f} €")
        elif tiene_cupon == "no":
            # No hay cupón, precio final igual al original
            print(f"El precio final es: {precio_original:.2f} €")
        else:
            print("Respuesta no válida. Debes responder 'sí' o 'no'.")
except ValueError:
    print("Precio inválido. Debes ingresar un número.")
```
