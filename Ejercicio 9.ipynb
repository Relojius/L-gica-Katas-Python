```python
# 9. Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista
# excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es ["Mapache", "Tigre",
# "Serpiente Pitón", "Cocodrilo", "Oso"].Usa la función filter()
def filtrar_mascotas(mascotas):
    prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
    # Filtramos las mascotas que NO están en la lista de prohibidas
    mascotas_permitidas = list(filter(lambda m: m not in prohibidas, mascotas))
    return mascotas_permitidas

# Ejemplo de uso:
lista_mascotas = ["Perro", "Gato", "Tigre", "Conejo", "Serpiente Pitón"]
resultado = filtrar_mascotas(lista_mascotas)

print("Mascotas permitidas:", resultado)
```
