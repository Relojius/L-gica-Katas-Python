```python
# 30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras
# pero en diferente orden.
def son_anagramas(palabra1, palabra2):
    # Convertir a minúsculas y eliminar espacios
    p1 = palabra1.replace(" ", "").lower()
    p2 = palabra2.replace(" ", "").lower()
    
    # Verificar si tienen las mismas letras con la misma frecuencia
    return sorted(p1) == sorted(p2)

# Pedir palabras al usuario
pal1 = input("Introduce la primera palabra: ")
pal2 = input("Introduce la segunda palabra: ")

# Comprobar si son anagramas
if son_anagramas(pal1, pal2):
    print("Son anagramas.")
else:
    print("No son anagramas.")
```
