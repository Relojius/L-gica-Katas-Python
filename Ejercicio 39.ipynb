```python
# 39. Escribe un programa que determine qué calificación en texto tiene un alumno en base a su calificación numérica.
# Pedir calificación al usuario
nota = input("Introduce la calificación del alumno (0-100): ")

try:
    nota = int(nota)
    if nota < 0 or nota > 100:
        print("Por favor, introduce un número entre 0 y 100.")
    else:
        if 0 <= nota <= 69:
            calificacion_texto = "insuficiente"
        elif 70 <= nota <= 79:
            calificacion_texto = "bien"
        elif 80 <= nota <= 89:
            calificacion_texto = "muy bien"
        else:  # 90-100
            calificacion_texto = "excelente"

        print(f"La calificación del alumno es: {calificacion_texto}")
except ValueError:
    print("Por favor, introduce un número válido.")
```
