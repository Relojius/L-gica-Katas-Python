```python
#1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias.
#de cada letra en la cadena. Los espacios no deben ser considerados.
def contar_frecuencias(cadena):
    # Convertimos la cadena a minúsculas para no diferenciar entre mayúsculas y minúsculas
    cadena = cadena.lower()
    frecuencias = {}
    
    for caracter in cadena:
        if caracter != " ":  # ignorar espacios
            if caracter in frecuencias:
                frecuencias[caracter] += 1
            else:
                frecuencias[caracter] = 1
                
    return frecuencias

texto = "Hola mundo"
resultado = contar_frecuencias(texto)
print(resultado)
```

```python
#5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por defecto es 5. La función debe calcular la media de los números en la lista 
# y determinar si la media es mayor o igual que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". 
# La función debe devolveruna tupla que contenga la media y el estado.
def evaluar_notas(lista_numeros, nota_aprobado=5):
    if not lista_numeros:  # evitar división por cero
        return (0, "sin notas")

    media = sum(lista_numeros) / len(lista_numeros)

    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"

    return (media, estado)

notas = [7, 5, 6, 8, 4, 3, 2, 1, 3]
resultado = evaluar_notas(notas)
print(resultado)
```

```python
# 6. Escribe una función que calcule el factorial de un número de manera recursiva.
def factorial(n):
    if n < 0:
        raise ValueError("El factorial no está definido para números negativos.")
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))
```

```python
# 10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una
# excepción personalizada y maneja el error adecuadamente.
# Definimos una excepción personalizada
class ListaVaciaError(Exception):
    pass

def calcular_promedio(numeros):
    if not numeros:  # si la lista está vacía
        raise ListaVaciaError("La lista está vacía. No se puede calcular el promedio.")
    return sum(numeros) / len(numeros)

# Ejemplo de uso
try:
    lista = [5, 8, 10, 2]
    promedio = calcular_promedio(lista)
    print(f"El promedio es: {promedio}")
except ListaVaciaError as e:
    print(f"Error: {e}")
finally:
    print("Fin del programa.")
```

```python
# 28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.
def primer_duplicado(lista):
    vistos = set()
    for elemento in lista:
        if elemento in vistos:
            return elemento
        vistos.add(elemento)
    return None  # Si no hay duplicados

# Ejemplo de uso
numeros = [3, 5, 1, 2, 5, 3]
print("Primer duplicado:", primer_duplicado(numeros))
```

```python
# 30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras
# pero en diferente orden.

def son_anagramas(palabra1, palabra2):
    # Convertir a minúsculas y eliminar espacios
    p1 = palabra1.replace(" ", "").lower()
    p2 = palabra2.replace(" ", "").lower()
    
    # Verificar si tienen las mismas letras con la misma frecuencia
    return sorted(p1) == sorted(p2)

# Pedir palabras al usuario
pal1 = input("Introduce la primera palabra: ")
pal2 = input("Introduce la segunda palabra: ")

# Comprobar si son anagramas
if son_anagramas(pal1, pal2):
    print("Son anagramas.")
else:
    print("No son anagramas.")
```

```python
# 34 y 35.Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son:
# crecer_tronco , nueva_rama , crecer_rama
class Arbol:
    def __init__(self):
        self.tronco = 1      # Tronco inicial de longitud 1
        self.ramas = []      # Lista vacía de ramas

    def crecer_tronco(self):
        """Aumenta la longitud del tronco en 1"""
        self.tronco += 1

    def nueva_rama(self):
        """Agrega una nueva rama de longitud 1"""
        self.ramas.append(1)

    def crecer_ramas(self):
        """Aumenta en 1 la longitud de todas las ramas existentes"""
        self.ramas = [rama + 1 for rama in self.ramas]

    def quitar_rama(self, posicion):
        """Elimina la rama en la posición indicada (0-index)"""
        if 0 <= posicion < len(self.ramas):
            self.ramas.pop(posicion)
        else:
            print("Posición inválida. No se pudo quitar la rama.")

    def info_arbol(self):
        """Devuelve información sobre el árbol"""
        info = {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitudes_ramas": self.ramas
        }
        return info


# ======= Caso de uso =======

# 1. Crear un árbol
mi_arbol = Arbol()

# 2. Hacer crecer el tronco del árbol una unidad
mi_arbol.crecer_tronco()

# 3. Añadir una nueva rama
mi_arbol.nueva_rama()

# 4. Hacer crecer todas las ramas del árbol una unidad
mi_arbol.crecer_ramas()

# 5. Añadir dos nuevas ramas
mi_arbol.nueva_rama()
mi_arbol.nueva_rama()

# 6. Retirar la rama situada en la posición 2 (0-index)
mi_arbol.quitar_rama(2)

# 7. Obtener información sobre el árbol
info = mi_arbol.info_arbol()
print(info)
```

```python
# 36. Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta
# corriente. Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y
# agregar dinero al saldo.
class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):
        """Retira dinero del saldo del usuario. Lanza error si no hay suficiente saldo o no tiene cuenta."""
        if not self.cuenta_corriente:
            raise Exception(f"{self.nombre} no tiene cuenta corriente para retirar dinero.")
        if cantidad > self.saldo:
            raise Exception(f"{self.nombre} no tiene suficiente saldo para retirar {cantidad}. Saldo actual: {self.saldo}")
        self.saldo -= cantidad
        print(f"{self.nombre} ha retirado {cantidad}. Saldo actual: {self.saldo}")

    def transferir_dinero(self, otro_usuario, cantidad):
        """Recibe dinero desde otro usuario. Lanza error si el otro usuario no puede realizar la transferencia."""
        if not otro_usuario.cuenta_corriente:
            raise Exception(f"{otro_usuario.nombre} no tiene cuenta corriente para transferir dinero.")
        if cantidad > otro_usuario.saldo:
            raise Exception(f"{otro_usuario.nombre} no tiene suficiente saldo para transferir {cantidad}. Saldo actual: {otro_usuario.saldo}")
        otro_usuario.saldo -= cantidad
        self.saldo += cantidad
        print(f"{otro_usuario.nombre} ha transferido {cantidad} a {self.nombre}.")
        print(f"Saldo de {otro_usuario.nombre}: {otro_usuario.saldo}, Saldo de {self.nombre}: {self.saldo}")

    def agregar_dinero(self, cantidad):
        """Agrega dinero al saldo del usuario."""
        self.saldo += cantidad
        print(f"{self.nombre} ha agregado {cantidad}. Saldo actual: {self.saldo}")


# ======= Caso de uso =======

# 1. Crear dos usuarios: Alicia y Bob
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)

# 2. Agregar 20 unidades de saldo a Bob
bob.agregar_dinero(20)  # Bob saldo: 70

# 3. Hacer una transferencia de 80 unidades desde Bob a Alicia
try:
    alicia.transferir_dinero(bob, 80)  # Esto lanzará error porque Bob tiene solo 70
except Exception as e:
    print("Error:", e)

# Corregimos la transferencia a 70, que es lo que tiene Bob
alicia.transferir_dinero(bob, 70)  # Bob saldo: 0, Alicia saldo: 170

# 4. Retirar 50 unidades de saldo a Alicia
alicia.retirar_dinero(50)  # Alicia saldo: 120
```

```python
# 37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras ,
# reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro
# de la función procesar_texto .

# Funciones auxiliares
def contar_palabras(texto):
    palabras = texto.split()
    return len(palabras)

def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    return texto.replace(palabra_original, palabra_nueva)

def eliminar_palabra(texto, palabra):
    return texto.replace(palabra, "")
def procesar_texto(texto, opcion, palabra_original=None, palabra_nueva=None):
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        if palabra_original is None or palabra_nueva is None:
            print("Error: Para 'reemplazar' necesitas palabra_original y palabra_nueva")
            return texto
        return reemplazar_palabras(texto, palabra_original, palabra_nueva)
    elif opcion == "eliminar":
        if palabra_original is None:
            print("Error: Para 'eliminar' necesitas especificar la palabra")
            return texto
        return eliminar_palabra(texto, palabra_original)
    else:
        print("Opción inválida. Usa 'contar', 'reemplazar' o 'eliminar'.")
        return texto
texto = input("Ingresa un texto: ")

# Contar palabras
resultado_contar = procesar_texto(texto, "contar")
print("Conteo de palabras:", resultado_contar)

# Reemplazar palabra "Hola" por "Saludos"
resultado_reemplazar = procesar_texto(texto, "reemplazar", "Hola", "Saludos")
print("Texto reemplazado:", resultado_reemplazar)

# Eliminar palabra "Python"
resultado_eliminar = procesar_texto(texto, "eliminar", "Python")
print("Texto con palabra eliminada:", resultado_eliminar)
```

```python
# 40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o
# "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).
import math

def calcular_area(figura, datos):
    """
    Calcula el área de una figura geométrica.
    
    Parámetros:
    - figura: "rectangulo", "circulo" o "triangulo"
    - datos: tupla con los datos necesarios:
        * rectángulo: (base, altura)
        * círculo: (radio,)
        * triángulo: (base, altura)
    """
    figura = figura.lower()
    
    if figura == "rectangulo":
        if len(datos) != 2:
            raise ValueError("Rectángulo necesita (base, altura)")
        base, altura = datos
        return base * altura
    
    elif figura == "circulo":
        if len(datos) != 1:
            raise ValueError("Círculo necesita (radio,)")
        radio = datos[0]
        return math.pi * radio**2
    
    elif figura == "triangulo":
        if len(datos) != 2:
            raise ValueError("Triángulo necesita (base, altura)")
        base, altura = datos
        return (base * altura) / 2
    
    else:
        raise ValueError("Figura no válida. Debe ser 'rectangulo', 'circulo' o 'triangulo'.")


# ===== Ejemplos de uso =====
print("Área rectángulo (4x5):", calcular_area("rectangulo", (4, 5)))   # 20
print("Área círculo (radio 3):", calcular_area("circulo", (3,)))       # 28.274...
print("Área triángulo (4x6):", calcular_area("triangulo", (4, 6)))     # 12
```

```python
# 41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el
# monto final de una compra en una tienda en línea, después de aplicar un descuento.

precio_original = input("Ingresa el precio original del artículo: ")

try:
    precio_original = float(precio_original)
    if precio_original < 0:
        print("El precio no puede ser negativo.")
    else:
        # 2. Preguntar si tiene cupón de descuento
        tiene_cupon = input("¿Tienes un cupón de descuento? (sí/no): ").strip().lower()

        if tiene_cupon == "sí" or tiene_cupon == "si":
            # 3. Solicitar valor del cupón
            cupon = input("Ingresa el valor del cupón de descuento: ")
            try:
                cupon = float(cupon)
                if cupon > 0:
                    precio_final = precio_original - cupon
                    if precio_final < 0:
                        precio_final = 0  # Evitar precio negativo
                    print(f"El precio final después del descuento es: {precio_final:.2f} €")
                else:
                    print("El valor del cupón debe ser mayor a 0. No se aplica descuento.")
                    print(f"El precio final es: {precio_original:.2f} €")
            except ValueError:
                print("Cupón inválido. No se aplica descuento.")
                print(f"El precio final es: {precio_original:.2f} €")
        elif tiene_cupon == "no":
            # No hay cupón, precio final igual al original
            print(f"El precio final es: {precio_original:.2f} €")
        else:
            print("Respuesta no válida. Debes responder 'sí' o 'no'.")
except ValueError:
    print("Precio inválido. Debes ingresar un número.")
```
