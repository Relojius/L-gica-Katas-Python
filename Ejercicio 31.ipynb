```python
# 31. Crea una función que solicite al usuario ingresar una lista de nombres y luego solicite un nombre para buscar en
# esa lista. Si el nombre está en la lista, se imprime un mensaje indicando que fue encontrado, de lo contrario, se
# lanza una excepción.
def buscar_nombre():
    # Solicitar lista de nombres separada por comas
    entrada = input("Ingresa una lista de nombres separados por comas: ")
    nombres = [n.strip() for n in entrada.split(",")]  # Limpiar espacios

    # Solicitar el nombre a buscar
    nombre_buscar = input("Ingresa el nombre que quieres buscar: ")

    # Verificar si está en la lista
    if nombre_buscar in nombres:
        print(f"El nombre '{nombre_buscar}' fue encontrado en la lista.")
    else:
        raise Exception(f"El nombre '{nombre_buscar}' NO se encuentra en la lista.")

# Llamar a la función
buscar_nombre()
```
