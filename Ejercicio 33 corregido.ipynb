```python
# 33. Crea una funci√≥n lambda que sume elementos correspondientes de dos listas dadas.
# Si las listas tienen la misma longitud:
sumar_listas = lambda lista1, lista2: [a + b for a, b in zip(lista1, lista2)]

# Ejemplo de uso:
l1 = [1, 2, 3]
l2 = [4, 5, 6]
print(sumar_listas(l1, l2))  # [5, 7, 9]
```
